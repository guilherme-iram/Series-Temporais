```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, b=5): return x + b

X = np.linspace(1, 10, 100)
y = f(X)

for i in range(len(y)):
    y[i] += (np.random.randn() * 2) * (-1 * np.random.randint(0, 2)) 


    
```

```python
plt.plot(X, y, )
```

```python
residuals = y - f(X)
residuals
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(f(X), residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valores Previstos')
plt.ylabel('Resíduos')
plt.title('Gráfico de Dispersão de Resíduos vs. Valores Previstos')
plt.show()
```

```python
import statsmodels.api as sm
_, p_value, _, _ = sm.stats.diagnostic.het_breuschpagan(residuals, sm.add_constant(f(X)))
print(f'Valor-p do teste de Breusch-Pagan: {p_value}')
print(f"Se p não é menor que 0.05, então aceita-se hipotese nula (homeostacedade não é violada)")
```

```python
import statsmodels.api as sm
import scipy.stats as stats

figure, ax =  plt.subplots(figsize=(15, 8))
sm.qqplot(residuals, line='45',ax=ax)

plt.title('Gráfico QQ dos Resíduos')
plt.show()
```

```python
_, p_value = stats.shapiro(residuals)
print(f'Valor-p do teste de Shapiro-Wilk: {p_value}')
print(f'Se p é menor que 0.05, então aceita-se hipótese nula - resíduo tem distribuição normal')
```
